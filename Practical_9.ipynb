```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
# Draw line chart graph using Python matplotlib library.
values1 = [10, 18, 26, 34, 42, 50, 58, 66, 74, 82]
# Taking Values for Graph Plotting
values2 = [8, 15, 22, 29, 36, 34, 50, 55, 64, 71]
plt.plot(range(500, 5500, 500), values1, c='g', lw=1, ls='--', marker='o')
# Making Changes in graphs
plt.plot(range(500, 5500, 500), values2, c='c', lw=2, ls='--', marker='o')
plt.legend(['Fixed', 'Variable'])
# Printing Graph
plt.show()
```
